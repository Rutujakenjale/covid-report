```python
import pandas as pd
data=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df=data.iloc[:-1,1:]
```

```python
df[df.state_name.str.contains('Maharashtra',case=False)]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
total_states=np.arange(len(df['state_name']))
```

```python
total_states
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['death'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['death'])) 
plt.xlabel('death')  
plt.title('Corona Virus Cases')  
plt.show()
```
